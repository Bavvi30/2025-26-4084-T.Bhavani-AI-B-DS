```python
#z-score,statistical inference,random variable model,central limit theorem,standardizations
import numpy as np
import math
population=np.array([10,20,30,40,50,60,70])
N=len(population)
print("\nPopulation Size(N):",N)
mu=np.mean(population)
print("\nPopulation Mean(μ):",mu)
sigma2=np.var(population,ddof=0)
print("\nPopulation Variance(σ²):",sigma2)
sigma=np.sqrt(sigma2)
sample=np.array([20,30,40,50])
n=len(sample)
print("\nSample Size(n):",n)
x_bar=np.mean(sample)
print("\nSample Mean(x̄):",x_bar)
s2=np.var(sample,ddof=1)
print("\nSample Variance:",s2)
s=np.sqrt(s2)
print("\nSample Standard Deviation(s):",s)
SE=sigma/math.sqrt(n)
print("\nStandard Error(σ/√n):",SE)
z=(x_bar-mu)/SE
print("\nZ-score:",z)
```

```python
#Hypothesis Code
import math
n=100
x=60
p_hat=x/n
SE=math.sqrt(p_hat*(1-p_hat)/n)
z=1.96
lower=p_hat-z*SE
upper=p_hat+z*SE
print("Sample proportion=",p_hat)
print("Confidence Interval=",(lower,upper))
p0=0.50
z_test=(p_hat-p0)/math.sqrt(p0*(1-p0)/n)
print("z value=",z_test)
if abs(z_test)>1.96:
  print("Reject Ho")
else:
  print("Do not reject Ho")
```
