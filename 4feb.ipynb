```python
#data manupulation
import pandas as pd
data = {"Name":["A","B","C"],
        "Age":[20,21,22],
        "Marks": [70,80,90]}
df=pd.DataFrame(data)
print("Original data:")
print(df)
print("Adding a new column bonus:")
df["Bonus"] = 5
print(df)
print("\nModifiying a column marks:")
df["Marks"] = df["Marks"] + 2
print(df)
print("\nCreating a new column based on other columns:")
df["Total"] = df["Marks"] + df["Bonus"]
print(df)
print("\nRenaming a column:")
df.rename(columns={"Total":"Final"},inplace = True )
print(df)
print("\nDroping a column:")
df.drop("Bonus",axis=1,inplace=True)
print(df)
print("\nDroping a row:")
df.drop(1,axis=0,inplace=True)
print(df)
print("\nChanging a particular value:")
df.loc[0,"Age"]=21
print(df)
```

```python
n=int(input("Enter number of elements:"))
arr=[]
for i in range(n):
    arr.append(int(input(f"Enter element{i+1}:")))
print("\n Original array:",arr)
for i in range(n-1):
  min_index=i
  for j in range(i+1,n):
    if arr[j]<arr[min_index]:
      min_index=j
  arr[i],arr[min_index]=arr[min_index],arr[i]
print("sorted array:",arr)
```

```python
# Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Example usage:
my_list = [64, 34, 25, 12, 22, 11, 90]
print("Original array:", my_list)
sorted_list = bubble_sort(my_list)
print("Sorted array:", sorted_list)
```

```python
#binary search
n=int(input("Enter number of elements:"))
arr=[]
for i in range(n):
  arr.append(int(input(f"Enter element{i+1}:")))
arr.sort()
print("\nSorted list:", arr)
key= int(input("Enter the element to be searched:"))
low = 0
high = n - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print(f"Element found at position {mid + 1}")
        found = True
        break
    elif arr[mid] < key:
          high = mid + 1
    else:
          high = mid - 1
if not found:
      print("Element not found")
```
